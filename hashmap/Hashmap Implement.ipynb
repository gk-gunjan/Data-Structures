```python
class MapNode:
    def __init__(self,key,value):
        self.key=key
        self.value=value
        self.next=None
        
class Map:
    
    def __init__(self):
        self.bucketSize=10
        self.buckets=[None for i in range(self.bucketSize)]
        self.count=0
        
    def size(self):
        return self.count
    #getting the size of array list size
    def getBucketIndex(self,hc):
        return (abs(hc)%(self.bucketSize))
    
    def getValue(self,key):
        hc=hash(key)
        index=getBucketIndex(hc)
        head=self.buckets[index]
        while head is not None:
            if head.key ==key:
                return head.value
            head=head.next
        return None
    
    def remove(self,key):
        hc=hash(key)
        index=self.getBucketIndex(hc)
        head=self.buckets[index]
        prev=None
        while head is not None:
            if head.key ==key:
                if prev is None:
                    self.buckets[index]=head.next
                else:
                    prev.next=head.next
                self.count-=1
                return head.value
            prev=head
            head=head.next
        return None
    
    def rehash(self):
        temp=self.buckets
        self.buckets=[None for i in range(2*self.bucketSize)]
        self.bucketSize=2*self.bucketSize
        self.count=0
        for head in temp:
            while head is not None:
                self.insert(self.key,self.value)
                head=head.next
                
                
    
    #insertig the value as key value
    def insert(self,key,value):
        
        #using hash inbuild hash function to create key
        hc=hash(key)
        
        #getting index from the above function
        index=self.getBucketIndex(hc)
        head=self.buckets[index] #storing it in head
        while head is not None: #now run a loop till head is not none
            if head.key==key:
                head.value=value
                return #inserting in the value using linked list if value are different then head =head.next
            head=head.next
        head=self.buckets[index]
        newNode=MapNode(key,value) #for new key to be inserted we create a new node and increment the count by 1
        newNode.next=head
        self.buckets[index]=newNode
        self.count+=1
        loadFactor=self.count/self.bucketSize
        if loadFactor >0.7:
            self.rehash()
m=Map()
m.insert('gunjan',20)
print(m.size())
m.insert('gunjan',21)
print(m.size())
m.insert('gunjan',13)
print(m.size())
```
